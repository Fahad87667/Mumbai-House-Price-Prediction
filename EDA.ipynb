```python
# Function to convert values to a common unit (e.g., convert "Lacs" to "Cr")
def convert_to_common_unit(row):
    if row['price_unit'] == 'L':
        return row['price'] / 100
    elif row['price_unit'] == 'Cr':
        return row['price']
    else:
        return None  # Handle other units if needed

# Create a new column 'merged_price' by applying the conversion function
data['Price'] = data.apply(convert_to_common_unit, axis=1)

# Drop the original 'price' and 'price_unit' columns
data = data.drop(['price', 'price_unit'], axis=1)

```

```python
data.head()
```

```python
# Function to convert values to another unit (e.g., convert "Cr" to "Lacs")
def convert_to_another_unit(value):
    return value * 100
```

```python
# Apply the conversion function to the 'merged_price' column
data['price'] = data['Price'].apply(convert_to_another_unit)
```

```python
# Drop the original 'merged_price' column if needed
data = data.drop(['Price'], axis=1)
```

```python
x = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.metrics import r2_score
from sklearn.pipeline import make_pipeline
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y, test_size=0.30, random_state=0)
```

```python
x_train.shape,x_test.shape
```

# Applying Linear Regression

```python
# Assuming you have a DataFrame named 'df' with columns 'type', 'region', 'status'
column_trans = make_column_transformer(
    (OneHotEncoder(sparse_output=False), ['type', 'region', 'status']),
    remainder='passthrough'
)
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipeline = make_pipeline(column_trans,scaler,lr)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred_lr = pipeline.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

# Applying lasso

```python
lasso = Lasso()
pipeline = make_pipeline(column_trans, scaler, lasso)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred_lasso = pipeline.predict(x_test)
```

```python
r2_score(y_test,y_pred_lasso)
```

# Applying Ridge

```python
ridge = Ridge()
pipeline = make_pipeline(column_trans, scaler, ridge)
```

```python
pipeline.fit(x_train,y_train)
```

```python
y_pred_ridge = pipeline.predict(x_test)
```

```python
r2_score(y_test,y_pred_ridge)
```
